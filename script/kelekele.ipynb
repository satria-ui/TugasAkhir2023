# RAVDESS

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Confusion matrix values
cm_values = np.array([[0.89, 0, 0.11, 0, 0, 0],
                      [0, 0.89, 0, 0, 0, 0.11],
                      [0, 0.10, 0.80, 0.10, 0, 0],
                      [0, 0.20, 0, 0.80, 0, 0],
                      [0, 0, 0, 0, 0.80, 0.20],
                      [0, 0.10, 0, 0, 0.30, 0.60]])

# Define class labels
labels = ['Angry', 'Fear', 'Disgust', 'Happy', 'Neutral', 'Sad']

# Create DataFrame from the confusion matrix values
confmatrix_df = pd.DataFrame(cm_values, index=labels, columns=labels)

# Plot confusion matrix using seaborn heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(confmatrix_df, annot=True, annot_kws={"size": 10}, fmt='.2f')

# Customize the plot
plt.title('Normalized Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')

plt.show()
```
